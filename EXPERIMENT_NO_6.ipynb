Evaluate line integrals using Green’s theorem

Aim

Evaluate the line integral of the vector field along a curve using Green’s theorem.

Algorithm

Step 1: Import the library Sympy, Numpy.

Step 2: Define the Vector Field.

Step 3: Define the region $D$
.

Step 4: Compute the double integral in polar coordinates.

Step 5: Compute the vector field at specific points.

Step 6: Evaluate the vector field at each grid point.

Step 7: Plot the vector field and the region $D$
.

Problem 1 Evaluate the line integral of the vector field $F(x,y)=(x^2-y,x+y^2)$
 along the positively oriented unit circle $C$
 defined by $x^2+y^2=1$
 using Green’s theorem.


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variables
x, y = sp.symbols('x y')

# Define the vector field components
P = x**2 - y
Q = x + y**2

# Compute the partial derivatives
dQ_dx = sp.diff(Q, x)
dP_dy = sp.diff(P, y)

# Apply Green's theorem: double integral of (dQ/dx - dP/dy) over the region D
integrand = dQ_dx - dP_dy

# Define the region D: the unit disk x^2 + y^2 <= 1
# We will use polar coordinates to simplify the integral

r, theta = sp.symbols('r theta')
x_polar = r * sp.cos(theta)
y_polar = r * sp.sin(theta)

# Change of variables: integrand in polar coordinates
integrand_polar = integrand.subs({x: x_polar, y: y_polar}) * r

# Define the limits for r and theta
r_limits = (r, 0, 1)
theta_limits = (theta, 0, 2 * sp.pi)

# Compute the double integral in polar coordinates
area_integral = sp.integrate(integrand_polar, r_limits, theta_limits)

# Display the result
area_integral
```

##Visualization of Vector field and redion D

```python
import numpy as np
import matplotlib.pyplot as plt


# Define the vector field components
def P(x, y):
    return x**2 - y

def Q(x, y):
    return x + y**2

# Create a grid of points
x_vals = np.linspace(-1.5, 1.5, 20)
y_vals = np.linspace(-1.5, 1.5, 20)
X, Y = np.meshgrid(x_vals, y_vals)

# Compute the vector field at grid points
U = P(X, Y)
V = Q(X, Y)
#define the unit circle
theta = np.linspace(0, 2 * np.pi,100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
#plot the vector field
plt.figure(figsize=(8,8))
plt.quiver(X, Y, U, V, color='blue', alpha=0.5)
plt.plot(circle_x, circle_y, color='red', linewidth=2)
#shade the area inside the unit circle
plt.fill(circle_x, circle_y, color='red', alpha=0.2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vector Field and unit circle')
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.grid(color='gray',linestyle='--',linewidth=0.5)
plt.axis('equal')
plt.show()
```
