#Experiment No.3

Evaluating gradient of a scalar field and visualise

**Case**Find the gradient of the scalar $f=x^2+y^2+z^2$ and plot the gradient field.

**Solution**

Evaluating the gradient of the scalar field

```python
import sympy as sp

#Define the symbols
x,y,z = sp.symbols('x,y,z')

#Define the scalar field
scalar_field=x**2 + y**2 + z**2 #Example scalar field
scalar_field
```

```python
#Compute the gradient
gradient=sp.Matrix([sp.diff(scalar_field,var) for var in (x,y,z)])
print("The gradient of the scalar field is")
gradient
```

#Plotting gradient of the scalar field

```python
import numpy as np
import matplotlib.pyplot as plt

#Define a grid of points where you wnt to evaluate the gradient
x_vals=np.linspace(-10,10,20)
y_vals=np.linspace(-10,10,20)
z_val=0 #For 2d visualisation in the z=0 plane

X,Y =np.meshgrid(x_vals,y_vals)

#Initialise arrays to store gradient components
U=np.zeros_like(X,dtype=float)
V=np.zeros_like(X,dtype=float)

#Evaluate the gradient at each point
for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    grad_evaluated=gradient.subs({x:X[i,j],y:Y[i,j],z:z_val})
    U[i,j]=float(grad_evaluated[0])
    V[i,j]=float(grad_evaluated[1])

#Plot the gradient vectors using matplotlib
fig,ax=plt.subplots()
ax.quiver(X,Y,U,V)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gradient vectors of the scalar field')
plt.show()
```
