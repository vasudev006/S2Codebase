**Experiment No.8**

Solve the homogenous linear D.E and visualise

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#Define the symbols
t=sp.symbols('t')
y=sp.Function('y')(t)

#Define the D.E
diff_eq=sp.Eq(y.diff(t,t)+4*y.diff(t)+4*y,0)

#Solve the differential equation
sol=sp.dsolve(diff_eq,y)

print("Gerneral solution:",sol)
```

```python
#Apply initial conditions
c1,c2=sp.symbols('C1 C2')
y_0=sol.rhs.subs({t:0})
eqn1=sp.Eq(y_0,1)
y_prime=sol.rhs.diff(t)
y_prime0=y_prime.subs({t:0})
eqn2=sp.Eq(y_prime0,-2)
solution=sp.solve((eqn1,eqn2),(c1,c2))
particular_sol=sol.rhs.subs(solution)
particular_solution=sp.Eq(y,particular_sol)
particular_solution
```
