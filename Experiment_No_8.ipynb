**Experiment No.8**



 solve the homogeneous linear differentiall equation and visualize


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
#define the symbols
t = sp.symbols('t')
y = sp.Function('y')(t)

#define the differential equation
diff_eq = sp.Eq(y.diff(t,t) + 4*y.diff(t)  + 4*y, 0)
#solve the differential eqaution
sol = sp.dsolve(diff_eq, y)
print("General solution:",sol)

```

```python
#Apply the intial conditions
c1, c2 = sp.symbols('C1 C2')
y_0=sol.rhs.subs({t:0})
eqn1=sp.Eq(y_0,1)
y_prime= sol.rhs.diff(t)
y_prime0=y_prime.subs({t:0})
eqn2=sp.Eq(y_prime0,-2)
solution=sp.solve((eqn1,eqn2),(c1,c2))
particular_sol = sol.rhs.subs(solution)
particular_solution =sp.Eq(y,particular_sol)
particular_solution
```

```python
#convert sympy solution toa numerical function
f = sp.lambdify(t, particular_sol, 'numpy')

#time range for plotting
t_vals = np.linspace(0,10,400)

#calculate numerical calculation
y_vals = f(t_vals)

#plot the solution
plt.plot(t_vals,y_vals, label='y(t)')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solution of y\'\' + 4y\'+4y = 0')
plt.legend()
plt.grid(True)
plt.show()
```

$$m \dfrac{d^2 x}{dt^2}+c \dfrac{dx}{dt}+kx=0$$

given $m=1,c=0.5,k=2$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
#define the symbols
t = sp.symbols('t')
x = sp.Function('x')(t)
#give valiues top the constants
m=1;c=0.5;k=2
#define the differential equation
diff_eq = sp.Eq(m*x.diff(t,t) + c*x.diff(t) + k*x, 0)
#solve the diffrential equation
sol = sp.dsolve(diff_eq, x)
print("General solution:",sol)
```

```python
#APPLY INTIAL CONDITIONS
c1,c2, = sp.symbols('C1 C2')
x_0=sol.rhs.subs({t:0})
eqn1=sp.Eq(x_0,1)
x_prime= sol.rhs.diff(t)
x_prime0=x_prime.subs({t:0})
eqn2=sp.Eq(x_prime0,0)
solution=sp.solve((eqn1,eqn2),(c1,c2))
particular_sol = sol.rhs.subs(solution)
particular_solution =sp.Eq(x,particular_sol)
particular_solution
```

```python
#convert sympy solution toa numerical function
f = sp.lambdify(t, particular_sol, 'numpy')

#time range for plotting
t_vals = np.linspace(0,10,400)

#calculate numerical calculation
x_vals = f(t_vals)

#plot the solution
plt.plot(t_vals,x_vals, label='x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Solution of x\'\' + 4x\'+4x = 0')
plt.legend()
plt.grid(True)
plt.show()
```
